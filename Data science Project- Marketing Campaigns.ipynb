```python
# created a function to observe the missing values
def missing_values_table(df):
        # Total missing values
        mis_val = df.isnull().sum()
        
        # Percentage of missing values
        mis_val_percent = 100 * df.isnull().sum() / len(df)
        
        # Make a table with the results
        mis_val_table = pd.concat([mis_val, mis_val_percent], axis=1)
        
        # Rename the columns
        mis_val_table_ren_columns = mis_val_table.rename(
        columns = {0 : 'Missing Values', 1 : '% of Total Values'})
        
        # Sort the table by percentage of missing descending
        mis_val_table_ren_columns = mis_val_table_ren_columns[
            mis_val_table_ren_columns.iloc[:,1] != 0].sort_values(
        '% of Total Values', ascending=False).round(1)
        
        # Print some summary information
        print ("Your selected dataframe has " + str(df.shape[1]) + " columns.\n"      
            "There are " + str(mis_val_table_ren_columns.shape[0]) +
              " columns that have missing values.")
        
        # Return the dataframe with missing information
        return mis_val_table_ren_columns
```

### Hypothesis test 

```python
# hypothesis test - 1

import scipy.stats as stats
from scipy.stats import chi2_contingency


# Null hypothesis: Older people are not as tech-savvy and probably prefer shopping in-store.

# At first find out the table of older people  (age > 60)
old_people = df_MarketingData[df_MarketingData['age'] > 60]

# Now find out the different catagorical purchase of old people in 
NumWebPurchases_OldPeople = sum(old_people['NumWebPurchases'])
NumCatalogPurchases_OldPeople = sum(old_people['NumCatalogPurchases'])
NumStorePurchases_OldPeople = sum(old_people['NumStorePurchases'])

# Then find out the table of young people (age < 60)
young_people = df_MarketingData[df_MarketingData['age'] < 60]

# Similarly find out the different catagorical purchase of young people
NumWebPurchases_YoungPeople = sum(young_people['NumWebPurchases'])
NumCatalogPurchases_YoungPeople = sum(young_people['NumCatalogPurchases'])
NumStorePurchases_YoungPeople = sum(young_people['NumStorePurchases'])


data = {
    'AgeGroup': ['OldPeople', 'YoungPeople'],
    'WebPurchases': [NumWebPurchases_OldPeople, NumWebPurchases_YoungPeople],
    'CatalogPurchases': [NumCatalogPurchases_OldPeople, NumCatalogPurchases_YoungPeople],
    'StorePurchases' : [NumStorePurchases_OldPeople, NumStorePurchases_YoungPeople]
}


# Create DataFrame
df = pd.DataFrame(data)

print('\n', df)

# Create a sepate column in df data frame for OnlinePurchases
df['OnlinePurchases'] = df['WebPurchases'] + df['CatalogPurchases']

# Drop the duplcate colums: WebPurchases and CatalogPurchases
df_Online_Store = df.drop(columns= ['WebPurchases', 'CatalogPurchases'])

# Perform the Chi squre test
contigency_table = pd.crosstab(index = df_Online_Store['AgeGroup'],  columns = [df_Online_Store['StorePurchases'],df_Online_Store['OnlinePurchases']])
chi2_stat, p_value, dof, chai_array = stats.chi2_contingency(contigency_table)

# Output results
print("\nChi-Square Statistic:", round(chi2_stat, 2))
print("Degrees of Freedom:", dof)
print('P value =', p_value)
print('chai_array = \n', chai_array)

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("\nResult: Reject the null hypothesis.")
    print("Conclusion: Older people are as tech friendly.")
else:
    print("\nResult: Fail to reject the null hypothesis.")
    print("Conclusion: Older people are not as tech-savvy and probably prefer shopping in-store.")
```
